```python
def primeFactors(n):
    f=2
    while n>int(n**0.5):
        if n%f == 0:
            yield f
            n = int(n/f)
        else:
            f = f+1
    

for x in primeFactors(1024):
    print(x, " ", end="")
```

```python
def genrange(*args):
    if len(args) == 0 or len(args) >3:
        print ("error")
        return
    elif len(args) == 1:
        b,a,n = args[0], 0, 1
    elif len(args) == 2:
        a,b,n = args[0], args[1], 1
    elif len(args) == 3:
        a,b,n = args[0], args[1], args[2]
        if a>b and  n>0:
            print ("error")
        elif a<b and n<0:
            print ("error")
            return
    v=a
    if n>0:
        while v<=b:
            yield v
            v = v+n
    elif n<0:
        while v>=b:
            yield v
            v=v+n
```

```python
for x in genrange(10):
    print(x)
```
