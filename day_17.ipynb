```python
import scipy.integrate as inte
import numpy as np
```

```python

fx = lambda x: 2*x*np.sin(x**2)
lx = 2
ux = 3
val, err = inte.quad(fx, lx, ux)
print(val, err)
```

```python
gx = lambda y, x: np.exp(-(x**2 +y**2)) * np.sin(x+y)
xu, xl = 2 , 3
yu, yl = lambda x: 2*x , lambda x: np.sin(x)/2
val, err = inte.dblquad(gx, xu, xl, yu, yl)
print(val , "\n", err)
```

```python
hx = lambda x, y, z: (np.cos(x*y*z)*np.exp((-x**2)-y**2-z**2))/(x+y+z)
zu, zl = -2, 5
yu, yl = 3 ,2
xu, xl = 2 ,3                      
val, err = inte.tplquad(hx, xl, xu, yl, yu, zl, zu)
print(val , "\n", err)
```

```python
from scipy.integrate import trapezoid, simpson
x = [0, 2, 4, 6, 8, 10, 12, 14]
y = [0.5, 1.2, 1.8, 2.1, 1.7, 1.1, 0.7, 0.1]
tp_val = trapezoid(x,y)
simp_val = simpson(x,y)
print(tp_val)
print(simp_val)
```

```python
from scipy.interpolate import interp1d
x = [0, 2, 4, 6, 8, 10, 12, 14]
y = [0.5, 1.2, 1.8, 2.1, 1.7, 1.1, 0.7, 0.1]

lp1 = interp1d (x,y, kind="linear")
lp2 = interp1d(x,y, kind="quadratic")
lp3 = interp1d(x,y, kind="cubic")

xs = np.linspace(0, 14, 15)
ys1 = lp1(xs)
ys2 = lp2(xs)
ys3 = lp3(xs)
fig ,ax = plt.subplots(figsize=(16,9))
plt.plot(x,y, "ok" ,label='base data')
plt.plot(xs,ys1,'m', label= 'linear fn')
plt.plot(xs,ys2,'y', label= 'quadratic fn')
plt.plot(xs,ys3,'g', label= 'cubic fn')
plt.fill_between(xs, ys3, 'g', alpha=0.2)
#plt.fill_between(xs, ys2, 'y', alpha=0.2)
#plt.fill_between(xs, ys3, 'm', alpha=0.2)
plt.legend()
plt.show()
```
