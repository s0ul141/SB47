```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def histogram_kernel(u):
    return 0.5 * (np.abs(u) <= 0.5)

def pdfHistogram(observations, x, bandwidth=2):
    n = len(observations)
    kde = np.sum([histogram_kernel((x - x_i) / bandwidth) 
                  for x_i in observations], axis=0)
    pdf = kde / (n * bandwidth)
    return pdf
```

```python
# Example usage
observations = [2, 2, 3, 4, 8, 9, 9, 10, 12]
x = np.linspace(min(observations) - 2, max(observations) + 2, 100)

# Visualize individual PDFs
plt.figure(figsize=(8, 6))
for x_i in observations:
    pdf = pdfHistogram([x_i], x)
    plt.plot(x, pdf, label=f"x = {x_i}")
plt.title("Individual Probability Density Functions (Histogram Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Visualize the total PDF
total_pdf = pdfHistogram(observations, x)
plt.figure(figsize=(8, 6))
plt.plot(x, total_pdf)
plt.title("Total Probability Density Function (Histogram Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```
