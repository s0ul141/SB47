```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from numpy import log
```

```python
df = pd.read_excel("Passenger.xlsx")
df
```

```python
n = int(input('Enter the number of years for Moving Average:'))
df['MA']=df['Passengers'].rolling(window = n).mean()
print((df['MA']).head(30))
```

```python
df['Passengers'].plot(color='green',linewidth =3,figsize=(12,6),label = 'Actual data')
df['MA'].plot(color='red',linewidth =3,figsize=(12,6),label = 'MA Values')
plt.legend()
plt.title('Passenger Value MA for 7 years')
plt.show()









```

```python
df['Passengers_log'] = np.log(df['Passengers'])
df['Passengers_log_diff2'] = df['Passengers_log']-df['Passengers_log'].shift(2)
result = adfuller(df['Passengers_log_diff2'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

Hence, our data has been transformed into stationary data.

```python
from statsmodels.tsa.stattools import adfuller,acf,pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
acf_val = acf(df['Passengers_log_diff2'].dropna())
```

```python
acf_val
```

```python
plot_acf(df['Passengers_log_diff2'].dropna(),lags=40)
```

```python
plot_pacf(df['Passengers_log_diff2'].dropna(),lags=40)
```
