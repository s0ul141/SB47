## Q1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def epanechnikov_kernel(u):
    return 0.75 * (1 - u**2) * (np.abs(u) <= 1)

def pdfEpanechnikov(observations, x, bandwidth=1.0):
    n = len(observations)
    kde = np.sum([epanechnikov_kernel((x - x_i) / bandwidth) 
                  for x_i in observations], axis=0)
    pdf = kde / (n * bandwidth)
    return pdf
```

```python
# Example usage
observations = [2, 2, 3, 4, 8, 9, 9, 10, 12]
x = np.linspace(min(observations) - 2, max(observations) + 2, 100)

# Visualize individual PDFs
plt.figure(figsize=(8, 6))
for x_i in observations:
    pdf = pdfEpanechnikov([x_i], x)
    plt.plot(x, pdf, label=f"x = {x_i}")
plt.title("Individual Probability Density Functions (Epanechnikov Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Visualize the total PDF
total_pdf = pdfEpanechnikov(observations, x)
plt.figure(figsize=(8, 6))
plt.plot(x, total_pdf)
plt.title("Total Probability Density Function (Epanechnikov Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

```python
observations = [2, 2, 3, 4, 8, 9, 9, 10, 12]
x = np.array([5])  # Evaluate the PDF at x = 5
pdf_at_5 = pdfEpanechnikov(observations, x)
print(f"PDF value at x = 5: {pdf_at_5[0]}")
```

## Q2

```python
def gaussian_kernel(u):
    """Gaussian kernel function."""
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * u**2)

def pdfGaussian(observations, x, bandwidth=1.0):
    n = len(observations)
    kde = np.sum([gaussian_kernel((x - x_i) / bandwidth) 
                  for x_i in observations], axis=0)
    pdf = kde / (n * bandwidth)
    return pdf
```

```python
# Example usage
observations = [2, 2, 3, 4, 8, 9, 9, 10, 12]
x = np.linspace(min(observations) - 2, max(observations) + 2, 100)

# Visualize individual PDFs
plt.figure(figsize=(8, 6))
for x_i in observations:
    pdf = pdfGaussian([x_i], x)
    plt.plot(x, pdf, label=f"x = {x_i}")
plt.title("Individual Probability Density Functions (Gaussian Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
pdf = pdfGaussian(observations, x)

# Visualize the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf)
plt.title("Probability Density Function (Gaussian Kernel)")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

```python
observations = [2, 2, 3, 4, 8, 9, 9, 10, 12]
x = np.array([5])  # Evaluate the PDF at x = 5
pdf_at_5 = pdfGaussian(observations, x)
print(f"PDF value at x = 5: {pdf_at_5[0]}")
```
