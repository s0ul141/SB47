```python
#question 1
A = np.array([[17, 24, 1, 8, 15],[23,5,7,14,16],[4,6,13,20,22],[10,12,19,21,3],[11,18,25,2,9]])
A
```

```python
sum_col= np.sum(A, axis=0)
sum_row=np.sum(A, axis=1)

diag_sum1= np.trace(A)
diag_sum2= np.trace(np.fliplr(A))

max_col_sum=np.max(sum_col)
max_row_sum=np.max(sum_row)
min_col_sum=np.min(sum_col)
min_col_sum=np.min(sum_row)

is_magic = (max_col_sum==max_row_sum == min_col_sum ==min_col_sum)
is_magic
```

```python
#Question 4
x=np.linspace(0, 10, 100)
y1 = ((np.exp(x)+np.sin(x))*(x/3))**0.5
y2 = (0.5*x)/(2+np.cos(x))
plt.plot(x, y1, label = "y= root of (exp(x)+sin(x)*(x/3))")
plt.plot(x, y2, label = "y= o.5x/2+cos(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#question 7
std_id = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10']
math = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
phy =  [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]
# i scatter plot 
plt.scatter(std_id, math, color = 'red')
plt.scatter(std_id, phy, color = 'green')
plt.xlabel("Student ID")
plt.ylabel("Marks in Maths & Physics")
plt.show()
```

```python
#Question 8
fig ,ax=plt.subplots(2,1, figsize=(16,9))
x = np.linspace(2*-np.pi, 2*np.pi, 1000)
y1 = np.sin(x**2)+np.cos(x)
y2 = (2*x)*(np.cos(x**2))-np.sin(x)
ax[0].plot(x,y1)
ax[1].plot(x,y2)

ax[0].spines['right'].set_color('none')
ax[0].spines['top'].set_color('none')
ax[1].spines['top'].set_color('none')

ax[0].xaxis.set_ticks_position('bottom')
ax[1].yaxis.set_ticks_position('left')

ax[0].spines['bottom'].set_position(('data', 0))
ax[1].spines['right'].set_position(('data', 0))
ax[0].spines['left'].set_position(('data', 0))

ax[1].spines['left'].set_position(('data', 0))
ax[1].spines['bottom'].set_position(('data', 0))

ax[1].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax[0].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])

plt.show()
fig.savefig('MyPlot.png')
```
