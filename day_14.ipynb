```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##ppstyle
x1=np.linspace(-np.pi,   np.pi,100)
y1=np.sin(x1)
y2=np.cos(x1)

plt.plot(x1,y1,':r', label='sin(x)') #-- for dashed, -. for dash dot, : ofr dotted
plt.plot(x1,y2,'-.k' ,label='cos(x)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("plots of $\sin(x)$ and $\cos(x)$")
plt.legend()
plt.show()
```

```python
x= np.arange(1,11)
y1=np.random.rand(10)
y2=np.random.rand(10)
y3=np.random.rand(10)
plt.bar(x,y1, color="m", label="y1")
plt.bar(x,y2, color="b", bottom=y1 ,label="y2" )
plt.bar(x,y3, color="g", label="y3", bottom=y1+y2)
plt.legend()
plt.show()
```

```python
x= np.arange(1,11)
y1=np.random.rand(10)
y2=np.random.rand(10)
y3=np.random.rand(10)
plt.barh(x,y1, color="m", label="y1")
plt.barh(x,y2, color="b", left=y1 ,label="y2" )
plt.barh(x,y3, color="g", label="y3", left=y1+y2)
plt.legend()
plt.show()
```

```python
x= np.arange(1,11)
y1=np.random.rand(10)
y2=np.random.rand(10)
y3=np.random.rand(10)
fig , ax=plt.subplots()
ax.bar(x,y1, color="m", bottom=y2,label="y1")
ax.bar(x,y2, color="b",  label="y2" )
ax.bar(x,y3, color="g", label="y3", bottom=y1+y2)
plt.xticks(x,["A","B","C","D","E","F","G","H","I","J"])
plt.legend()
plt.show()
```

```python
x=np.linspace(-5,5,100)
y1=np.sin(x)
y2=np.cos(x)
fig, ax= plt.subplots(figsize=(10,5))
ax.plot(x,y1)
ax.plot(x,y2)
ax.fill_between(x,y1,y2, where =(y1>=y2),color="g", alpha=0.3)
ax.fill_between(x,y1,y2, where =(y1<=y2),color="g", alpha=0.6)
plt.show()
```

```python
x=np.linspace(-2,2,1000)
y1=np.cos(40*x)*np.exp(-x**2)
y2=np.exp(-x**2)
y3=-np.exp(-x**2)
fig, ax= plt.subplots(figsize=(12,7))
ax.plot(x,y1, color="r", label="cos(x)^e^(-x^2)")
ax.plot(x,y2,"--b",label="e^-x^2")
ax.plot(x,y3,"--", color="b", label="-e^-x^(-2)")
#ax.fill_between(x,y1,y2, where =(y1>=y2),color="g", alpha=0.3)
#ax.fill_between(x,y1,y2, where =(y3<=y2),color="g", alpha=0.6)
plt.legend()
plt.show()
```
